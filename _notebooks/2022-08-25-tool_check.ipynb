# Conda Packages

```bash
conda list
```

```bash
echo Conda Check
# test for a kernel installation
test="jupyter" # keyword
check=`conda list | grep $test` # run command
n=${#check} # determine length

if [[ ${n} > 0 ]];  # testt length
then # greater than zero
    echo "$check"
else # less than zero
    echo "$test not found"
fi
```

### Jupyter Checks

```bash
jupyter --version
jupyter kernelspec list
```

```bash
# test for a kernel installation
test="python3" # keyword
check=`jupyter kernelspec list | grep $test` # run command
n=${#check} # determine length

if [[ ${n} > 0 ]];  # testt length
then # greater than zero
    echo "$check"
else # less than zero
    echo "$test not found"
fi
```
